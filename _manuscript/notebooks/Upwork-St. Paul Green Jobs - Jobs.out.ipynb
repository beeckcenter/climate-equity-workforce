```python
import pandas as pd
```

```python
# o_net_green_jobs_df = pd.read_csv('/content/drive/MyDrive/Upwork - St. Paul Clean Energy Jobs/The National Center for O_NET Development - Construction - Sheet1.csv')
o_net_green_jobs_df = pd.read_csv('/content/drive/MyDrive/Upwork - St. Paul Clean Energy Jobs/st_paul_green_jobs_data/raw_data/The National Center for O*NET Development/Green Enhanced Skills Occupations - All - Sheet1.csv')
uc_clean_energy_jobs_df = pd.read_csv('/content/drive/MyDrive/Upwork - St. Paul Clean Energy Jobs/clean_energy_job_quality_education_data.xlsx - Sheet 1.csv')
bls_education_df = pd.read_csv('/content/drive/MyDrive/Upwork - St. Paul Clean Energy Jobs/BLS - Employment Projection - Education.xlsx - Education and training assignments by detailed occupation.csv')
```

```python
# o_net_green_jobs_df.head()
o_net_green_jobs_df['O*NET-SOC Title'] = o_net_green_jobs_df['O*NET-SOC Title'].str.strip()
o_net_green_jobs_df['O*NET-SOC Code'] = o_net_green_jobs_df['O*NET-SOC Code'].str.split('.').str[0]
# o_net_green_jobs_df['O*NET-SOC Code'] = o_net_green_jobs_df['O*NET-SOC Code'].str.replace('-', '')
```

```python
green_jobs_uc_df_merged = pd.merge(o_net_green_jobs_df, uc_clean_energy_jobs_df, left_on='O*NET-SOC Title', right_on='occ_title', how='left')
```
